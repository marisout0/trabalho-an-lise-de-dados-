```python
#exercício de sala 1
#quanto mais dados, mais próxima a precisão e aproximação
#analisar esse código e o que ele faz pra vê se é coerente com o que você quer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Generate 100 synthetic data points for linear regression
np.random.seed(0)
X = 2 * np.random.rand(1000, 1)
y = 0.2 + 5 * X + np.random.randn(1000, 1)
# o ruído é os números aleatórios np.random.randn(100, 1)
# 4 + 3 * X é a equação linear (isso pode ser alterado)

# Create a pandas DataFrame
data = pd.DataFrame({'X': X.flatten(), 'y': y.flatten()})

# Perform linear regression - primeiro você define o objeto e depois as ações
model = LinearRegression()
#toda a análise da regressão ta dentro do comando LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Create a scatter plot of the generated data and the regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='y', data=data, label='Dados Gerados')
plt.plot(X, y_pred, color='red', linewidth=2, label='Reta de Regressão')
plt.title('Dados Gerados para Regressão Linear com Reta de Regressão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#exercício de aula 2 gráfico gadient descents

import numpy as np
import matplotlib.pyplot as plt

# Generate some data (simple linear relationship with noise)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term to X
X_b = np.c_[np.ones((100, 1)), X]

# Set hyperparameters
learning_rate = 0.1
n_iterations = 100
m = 100 # number of data points

# Initialize random weights (theta)
theta = np.random.randn(2, 1)

# Store cost at each iteration
cost_history = []

# Gradient Descent
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
    # Calculate and store the cost (Mean Squared Error)
    cost = np.mean((X_b.dot(theta) - y)**2)
    cost_history.append(cost)

# Plot the cost function over iterations
plt.figure(figsize=(10, 6))
plt.plot(range(n_iterations), cost_history)
plt.xlabel('Número de Iterações')
plt.ylabel('Custo (MSE)')
plt.title('Descida do Gradiente: Custo vs. Iterações')
plt.grid(True)
plt.show()

print("Final theta (intercept and coefficient):", theta)
```

# Task

## Gerar dados

### Subtask:
Crie conjuntos de dados sintéticos de tamanhos diferentes para a regressão linear.


**Reasoning**:
The subtask requires generating synthetic datasets of different sizes for linear regression. This involves creating numpy arrays for features (X) and target (y) based on a linear equation with added noise, following the specified sizes.



```python
# Generate synthetic data for different sizes

# Small dataset (100 points)
X_small = 2 * np.random.rand(100, 1)
y_small = 0.2 + 5 * X_small + np.random.randn(100, 1)

# Medium dataset (500 points)
X_medium = 2 * np.random.rand(500, 1)
y_medium = 0.2 + 5 * X_medium + np.random.randn(500, 1)

# Large dataset (1000 points)
X_large = 2 * np.random.rand(1000, 1)
y_large = 0.2 + 5 * X_large + np.random.randn(1000, 1)
```

## Treinar modelos

### Subtask:
Treine o modelo de regressão linear em cada conjunto de dados.


**Reasoning**:
The task is to train a linear regression model on each dataset. This involves importing the necessary class, instantiating models, and fitting them to the respective data.



```python
from sklearn.linear_model import LinearRegression

# Instantiate LinearRegression objects for each dataset size
model_small = LinearRegression()
model_medium = LinearRegression()
model_large = LinearRegression()

# Fit each model to its corresponding dataset
model_small.fit(X_small, y_small)
model_medium.fit(X_medium, y_medium)
model_large.fit(X_large, y_large)
```

## Fazer previsões

### Subtask:
Use os modelos treinados para fazer previsões nos dados originais ou em um conjunto de teste.


**Reasoning**:
Use the trained models to make predictions on the corresponding datasets.



```python
# Make predictions using the trained models
y_pred_small = model_small.predict(X_small)
y_pred_medium = model_medium.predict(X_medium)
y_pred_large = model_large.predict(X_large)
```

## Avaliar o desempenho

### Subtask:
Calcule métricas de desempenho (como o Erro Quadrático Médio - MSE) para cada modelo.


**Reasoning**:
Calculate the Mean Squared Error for each model and print the results.



```python
from sklearn.metrics import mean_squared_error

# Calculate the Mean Squared Error for each model
mse_small = mean_squared_error(y_small, y_pred_small)
mse_medium = mean_squared_error(y_medium, y_pred_medium)
mse_large = mean_squared_error(y_large, y_pred_large)

# Print the calculated MSE values
print(f"MSE for small dataset: {mse_small}")
print(f"MSE for medium dataset: {mse_medium}")
print(f"MSE for large dataset: {mse_large}")
```

## Visualizar resultados

### Subtask:
Crie gráficos para comparar as retas de regressão e, opcionalmente, as métricas de desempenho para cada tamanho de conjunto de dados.


**Reasoning**:
Create scatter plots for each dataset size with their respective regression lines and a bar chart for MSE comparison.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create DataFrames for plotting
df_small = pd.DataFrame({'X': X_small.flatten(), 'y': y_small.flatten(), 'y_pred': y_pred_small.flatten()})
df_medium = pd.DataFrame({'X': X_medium.flatten(), 'y': y_medium.flatten(), 'y_pred': y_pred_medium.flatten()})
df_large = pd.DataFrame({'X': X_large.flatten(), 'y': y_large.flatten(), 'y_pred': y_pred_large.flatten()})

# Plotting the scatter plots and regression lines
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.scatterplot(x='X', y='y', data=df_small, ax=axes[0], label='Dados Gerados')
axes[0].plot(df_small['X'], df_small['y_pred'], color='red', linewidth=2, label='Reta de Regressão')
axes[0].set_title('Conjunto de Dados Pequeno')
axes[0].set_xlabel('X')
axes[0].set_ylabel('y')
axes[0].legend()
axes[0].grid(True)

sns.scatterplot(x='X', y='y', data=df_medium, ax=axes[1], label='Dados Gerados')
axes[1].plot(df_medium['X'], df_medium['y_pred'], color='red', linewidth=2, label='Reta de Regressão')
axes[1].set_title('Conjunto de Dados Médio')
axes[1].set_xlabel('X')
axes[1].set_ylabel('y')
axes[1].legend()
axes[1].grid(True)

sns.scatterplot(x='X', y='y', data=df_large, ax=axes[2], label='Dados Gerados')
axes[2].plot(df_large['X'], df_large['y_pred'], color='red', linewidth=2, label='Reta de Regressão')
axes[2].set_title('Conjunto de Dados Grande')
axes[2].set_xlabel('X')
axes[2].set_ylabel('y')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()

# Plotting the MSE comparison
mse_values = {'Pequeno': mse_small, 'Médio': mse_medium, 'Grande': mse_large}
mse_df = pd.DataFrame(list(mse_values.items()), columns=['Tamanho do Conjunto de Dados', 'MSE'])

plt.figure(figsize=(8, 6))
sns.barplot(x='Tamanho do Conjunto de Dados', y='MSE', data=mse_df)
plt.title('Comparação de MSE por Tamanho do Conjunto de Dados')
plt.xlabel('Tamanho do Conjunto de Dados')
plt.ylabel('Erro Quadrático Médio (MSE)')
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term
X_b = np.c_[np.ones((100, 1)), X]

# Set hyperparameters
learning_rate = 0.1
n_iterations = 10 # Reduced iterations to show intermediate steps
m = 100

# Initialize random weights
theta = np.random.randn(2, 1)

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.6, label='Dados Gerados')

# Perform Gradient Descent and plot intermediate lines
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients

    # Plot the regression line at some intermediate steps
    if iteration % 2 == 0: # Plot every 2 iterations
        y_pred = X_b.dot(theta)
        plt.plot(X, y_pred, linestyle='--', label=f'Iteração {iteration}')

# Plot the final regression line
y_final = X_b.dot(theta)
plt.plot(X, y_final, color='red', linewidth=2, label='Reta Final')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Gradiente Descendente: Passos Intermediários')
plt.legend()
plt.grid(True)
plt.show()

print("Final theta:", theta)
```
